# Working sith kernel PCA

## Moon-like shapes

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
X_moons, y_moons = make_moons(n_samples=100, random_state=123)
plt.scatter(X_moons[y_moons==0, 0], X_moons[y_moons==0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X_moons[y_moons==1, 0], X_moons[y_moons==1, 1], color='blue', marker='o', alpha=0.5)
plt.show()
```

## How classic PCA will perform?

```python
from sklearn.decomposition import PCA
import numpy as np
scikit_pca = PCA(n_components=2)
X_spca_moons = scikit_pca.fit_transform(X_moons)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7, 3))
ax[0].scatter(X_spca_moons[y_moons==0, 0], X_spca_moons[y_moons==0, 1], color='red', marker='^', alpha=0.5)
ax[0].scatter(X_spca_moons[y_moons==1, 0], X_spca_moons[y_moons==1, 1], color='blue', marker='o', alpha=0.5)
ax[1].scatter(X_spca_moons[y_moons==0, 0], np.zeros((50, 1))+0.02, color='red', marker='^', alpha=0.5)
ax[1].scatter(X_spca_moons[y_moons==1, 0], np.zeros((50, 1))-0.02, color='blue', marker='o', alpha=0.5)
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_ylim([-1, 1])
ax[1].set_yticks([])
ax[1].set_xlabel('PC1')
plt.show()
```

## Let's see how our implemented kernel PCA will do in this case (+gamma experiments):

```python
from kernelPCA import rbf_kernel_pca
from ipywidgets import interact, widgets, IntSlider
def visualize(g=15):
    X_kpca_moons, _ = rbf_kernel_pca(X_moons, gamma=g, n_components=2)
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7, 3))
    ax[0].scatter(X_kpca_moons[y_moons==0, 0], X_kpca_moons[y_moons==0, 1], color='red', marker='^', alpha=0.5)
    ax[0].scatter(X_kpca_moons[y_moons==1, 0], X_kpca_moons[y_moons==1, 1], color='blue', marker='o', alpha=0.5)
    ax[1].scatter(X_kpca_moons[y_moons==0, 0], np.zeros((50, 1))+0.02, color='red', marker='^', alpha=0.5)
    ax[1].scatter(X_kpca_moons[y_moons==1, 0], np.zeros((50, 1))-0.02, color='blue', marker='o', alpha=0.5)
    ax[0].set_xlabel('PC1')
    ax[0].set_ylabel('PC2')
    ax[1].set_ylim([-1, 1])
    ax[1].set_yticks([])
    ax[1].set_xlabel('PC1')
    plt.show()
    
interact(visualize, g=IntSlider(min=1, max=30, value=15, step=1, continuous_update=False))
```

## Concentric circles case

```python
from sklearn.datasets import make_circles
X_circles, y_circles = make_circles(n_samples=1000, random_state=123, noise=0.1, factor=0.2)
plt.scatter(X_circles[y_circles==0, 0], X_circles[y_circles==0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X_circles[y_circles==1, 0], X_circles[y_circles==1, 1], color='blue', marker='o', alpha=0.5)
plt.show()
```

## Classic PCA:

```python
scikit_pca = PCA(n_components=2)
X_spca_circles = scikit_pca.fit_transform(X_circles)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7, 3))
ax[0].scatter(X_spca_circles[y_circles==0, 0], X_spca_circles[y_circles==0, 1], color='red', marker='^', alpha=0.5)
ax[0].scatter(X_spca_circles[y_circles==1, 0], X_spca_circles[y_circles==1, 1], color='blue', marker='o', alpha=0.5)
ax[1].scatter(X_spca_circles[y_circles==0, 0], np.zeros((500, 1))+0.02, color='red', marker='^', alpha=0.5)
ax[1].scatter(X_spca_circles[y_circles==1, 0], np.zeros((500, 1))-0.02, color='blue', marker='o', alpha=0.5)
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_ylim([-1, 1])
ax[1].set_yticks([])
ax[1].set_xlabel('PC1')
plt.show()
```

## Kernel PCA (+gamma experiments):

```python
from kernelPCA import rbf_kernel_pca
from ipywidgets import interact, widgets, IntSlider
def visualize(g=15):
    X_kpca_circles, _ = rbf_kernel_pca(X_circles, gamma=g, n_components=2)
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(7, 3))
    ax[0].scatter(X_kpca_circles[y_circles==0, 0], X_kpca_circles[y_circles==0, 1], color='red', marker='^', alpha=0.5)
    ax[0].scatter(X_kpca_circles[y_circles==1, 0], X_kpca_circles[y_circles==1, 1], color='blue', marker='o', alpha=0.5)
    ax[1].scatter(X_kpca_circles[y_circles==0, 0], np.zeros((500, 1))+0.02, color='red', marker='^', alpha=0.5)
    ax[1].scatter(X_kpca_circles[y_circles==1, 0], np.zeros((500, 1))-0.02, color='blue', marker='o', alpha=0.5)
    ax[0].set_xlabel('PC1')
    ax[0].set_ylabel('PC2')
    ax[1].set_ylim([-1, 1])
    ax[1].set_yticks([])
    ax[1].set_xlabel('PC1')
    plt.show()
    
interact(visualize, g=IntSlider(min=1, max=30, value=15, step=1, continuous_update=False))
```

## Experiment with hipothetical new data (this time the projection will be on 1d subspace)

```python
X_moons_2, y_moons_2 = make_moons(n_samples=100, random_state=123)
alphas, lambdas = rbf_kernel_pca(X_moons_2, gamma=15, n_components=1)

x_new = X_moons_2[25]
print('new observation:', x_new)

x_proj = alphas[25]
print('x_proj:', x_proj)
```

```python
from kernelPCA import project_x
x_reproj = project_x(x_new, X_moons_2, gamma=15, alphas=alphas, lambdas=lambdas)
print('x_reproj:', x_reproj)
```

## Let's visualize the new projection to 1d subspace:

```python
plt.scatter(alphas[y_moons_2==0], np.zeros((50)), color='red', marker='^', alpha=0.5)
plt.scatter(alphas[y_moons_2==1], np.zeros((50)), color='blue', marker='o', alpha=0.5)
plt.scatter(x_proj, 0, color='green', marker='^', label='Primary projection of the X[25] observation', s=100)
plt.scatter(x_reproj, 0, color='black', marker='x', label='Reprojection of the X[25] observation', s=500)
plt.legend(scatterpoints=1)
plt.show()
```

# Using scikit-learn for kernel PCA

```python
from sklearn.decomposition import KernelPCA
X_skl, y_skl = make_moons(n_samples=100, random_state=123)
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_sklkpca = scikit_kpca.fit_transform(X_skl)

plt.scatter(X_sklkpca[y_skl==0, 0], X_sklkpca[y_skl==0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X_sklkpca[y_skl==1, 0], X_sklkpca[y_skl==1, 1], color='blue', marker='o', alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
