# Getting the dataset

```python
from sklearn import datasets
import numpy as np
iris = datasets.load_iris()
X = iris.data[:, [2,3]]
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

# Building a model

```python
from sklearn.neighbors import KNeighborsClassifier
knn_min = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')
knn_min.fit(X_train_std, y_train)
knn_man = KNeighborsClassifier(n_neighbors=5, p=2, metric='manhattan')
knn_man.fit(X_train_std, y_train)

import my_plotter
from my_plotter import plot_regions

plot_regions(x=X_train_std, y=y_train, clsf=knn_man, xl=my_plotter.iris_xl, yl=my_plotter.iris_yl, tit='kNN in Iris dataset (Manhattan metric)')
plot_regions(x=X_train_std, y=y_train, clsf=knn_min, xl=my_plotter.iris_xl, yl=my_plotter.iris_yl, tit='kNN in Iris dataset (Minkowski metric)')
```
